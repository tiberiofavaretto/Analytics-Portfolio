```python
# importing the random package
import random 

# Introduction to the game

"""
The following game is based on the episode 2x5 of the tv series Rick and Morty.
It requires knowing that Rolling Stones are a rock band and fun facts about numbers between one and ten.

Round 1: Understanding what the head wants.
Round 2: Knowing the favourite kind of music of the head.
Round 3: Guessing the number of verses.
Round 4: Distracting the opponent.

Known bugs and errors: None.

"""

# Creation of the first function
def introduction():
    print("""
A huge and threatening head appears in the sky and keep saying 'Show me what you got'. Fortunately you are the
smartest guy in the galaxy. Your name is Rick Sanchez and you, and your nephew Morty, are always in search of new
adventures.
Are you ready to save our planet?
    """)
    
    # addressing to the next function
    input("<Press enter to continue>\n")
    start()
    
# creation of function start
def start():
    print("""
You have also noticed that the giant head is wearing headphones.\n
What do you think the head wants? What does 'Show me what you got' mean?\n
    
A) It wants the treasures of earth, like gold or other valuable things.
B) It wants to see the greatest american football game ever played.
C) It wants to listen a hit song, you are in a singing competition.
    
    """)
    
    # Saving the answer into a variable
    head = input(prompt = "Make your choice.\n")
    
    # Conditional statement
    if head.lower() == 'a' or 'treasures' in head.lower() or 'gold' in head.lower() or 'first' in head.lower() or head == '1':
        print("""
        Heads don't care about gold!
        """)
        # link to the fail function
        input(prompt = "Press enter to see your result.\n")
        fail()
        
    elif head.lower() == 'b' or 'football' in head.lower() or 'second' in head.lower() or head == '2':
        print("""
        I would love to see the greatest american football game ever played, but the head doesn't want that.
        Maybe it is a Detroit Lions fan and has enough of football.
        """)
        
        # giving another chance to the user
        input(prompt = "<Press enter to restart>\n")
        start()
        
    elif head.lower() == 'c' or 'song' in head.lower() or 'hit' in head.lower() or 'third' in head.lower() or head == '3':
        print("""
             Yes, the head wants a hit song from earth, you really are the smartest guy on the planet!
        """)
        # going to the next stage
        input(prompt = "<Press enter to move on the next question>\n")
        begin()
        
    else:
        # wrong input, the level starts again
        print("Something went wrong.")
        input(prompt = "<Press enter to try again>\n")
        start()
    # end of conditional statement

# Creating the function for the second question
def begin():
    print("""
All the musicians are dead during the Grammy's Award night and you are the last hope for earth, so
the president ask to you and your nephew Morty to create a hit song for saving the planet.\n
Be careful! The head has a shirt that says 'I hate the Rolling Stones'.
What kind of music do you think the head would appreciate the most?
         
A) Classic music, heads are sofisticated.\n
B) Rock and roll! Heads just want to have fun.\n
C) Something alternative, maybe with meaningless words.\n
         
    """)
    
    # Creating the variable containing the answer of the user
    music = input(prompt = ">  \n")
    
    # Conditional statement based on variable choice
    if music.lower() == 'a' or 'classic' in music.lower() or 'sofisticated' in music.lower() or 'first' in music.lower() or music == '1':
        print("""
        That is not completely correct, but I like you and I will give you another chance.
        """)
        # giving another chance to the user
        input(prompt = "<Press enter to have another chance>\n")
        begin()
        
    elif music.lower() == 'b' or 'rock' in music.lower() or 'fun' in music.lower() or 'second' in music.lower() or music == '2':
        print("""
        Not even close!
        """)
        # addressing the user to the fail function
        input(prompt = "<Press enter to continue>\n")
        fail()
        
    elif music.lower() == 'c' or 'alternative' in music.lower() or 'meaningless' in music.lower() or 'third' in music.lower() or music == '3':
        print("""
        That is correct! You and your nephew compose a new song called 'Get Schwifty'.\n
        """)
        # moving on to the next level
        input(prompt = "<Press enter to move on>\n")
        scenario_C()
        
    else:
        # wrong input, the level starts again
        print("Something went wrong.")
        input(prompt = "<Press enter to try again>\n")
        begin()
        
    # end of the conditional statement
        
# Creating the function for the correct second choice
def scenario_C():
    
    # printing the options for the user
    print("""
How many verses should the hit song be composed of? (1-5)\n
You have three attempts.
    """)
    
    # defining number of chances
    # asks specifically for a number
    chances = 3
    
    # the optimal number is a random integer between 1 and 5
    perfection = random.randint(1, 5)
    
    # while loop
    while chances > 0:
        
        # asking to input the number of verses
        verses = input(prompt = "> ")
        
        # converting into an integer
        try:
            verses = int(verses)
        
        # impossibility of converting it
        except:
            # if the system couldn't convert it returns the following message
            print ("""
You have entered an invalid value. You have to restart the level.\n
Please insert a number, i.e. 1, 2, 3 and so on.
            """)
            
            # restarting the level
            input(prompt = "<Press enter if you got it>\n")
            scenario_C()
            
        else:
            # conditional statement
            if verses == perfection:
                print("""
                You are absolutely right! This is the right amount of verses.
                """)
                # moving on to the final stage
                input(prompt = "<Press enter to move on>\n")
                final()
                break
            
            elif verses != perfection:
            
                # reducing the chances
                chances -= 1
            
                # nested conditional based on the number of chances remaining
                if chances == 0:
                    # if the user used all the three guesses
                    print("\n You run out of guesses!")
                    # addresing the user to the fail function
                    input(prompt = "<Press enter to see your result>\n")
                    fail()
                    break
                
                # hints to the user
                elif verses > perfection:
                    print("Nope. Try again. Your guess was too high.")
                    
                elif verses < perfection:
                    print("Nope. Try again. Your guess was too low.")
                
            # end of nested conditional
            
    # end of the while loop
            
# Final question function
def final():
    
    print("""
It is not the end! Now you are at the finals. Your opponent is highly motivated to win.
Morty suggests distracting them. Which method is the most effective?\n
Hint: your opponent is a mormon with a very healthy lifestyle, not so used to drinking.\n
    
A) Dance the macarena.
B) Turn yourself into a pickle.
C) Put into their drinks a few drops of alcohol.
    
    """)
    
    # Variable distract
    distract = input(prompt = "<Make your choice>\n")
    
    # Conditional statement
    if distract.lower() == 'a' or 'dance' in distract.lower() or 'macarena' in distract.lower() or 'first' in distract.lower() or distract == '1':
        # link to the fail function
        input(prompt = "<Press enter to see your result>\n")
        fail()
    
    elif distract.lower() == 'b' or 'pickle' in distract.lower() or 'second' in distract.lower() or distract == '2':
        # link to the fail function
        input(prompt = "<Press enter to see your result>\n")
        fail()
        
    elif distract.lower() == 'c' or 'alcohol' in distract.lower() or 'drops' in distract.lower() or 'third' in distract.lower() or distract == '3':
        # correct answer, link to the final question
        print("""
How many of them then? (1-10)\n
You have two attempts.
        """)
        
        # defining number of attempts
        # asks specifically for a number
        attempts = 2
    
        # the optimal number is a random integer between 1 and 10
        exact = random.randint(1, 10)
        
        # while loop
        while attempts > 0:
        
            # try and except function
            try:
                # converting into integer
                drops = int(input(prompt = "Insert the number of drops.\n"))
    
            # failure of conversion
            except:
                # if the system couldn't convert it returns the following message
                print ("""
You have entered an invalid value. You have to restart the level.\n
Please insert a number, i.e. 1, 2, 3 and so on.
                """)
            
                # restarts the level
                input(prompt = "<Press enter if you got it.>\n")
                final()
                break
            
            else:
                if drops == exact:
                    print("It is the perfect number!")
                    # going to the win function
                    input(prompt = "<Press enter to see your result>\n")
                    win()
                    break
                    
                elif drops > 10 or drops < 1:
                    # the input is outside the range, the level starts again
                    print("Something went wrong. You need to insert a number between 1 and 10. You have to restart the level.")
                    input(prompt = "<Press enter if you got it.>\n")
                    final()
                    break
                
                elif drops != exact:
                    # reducing the chances
                    attempts -= 1
            
                    # nested conditional based on the number of chances remaining
                    # if the user used all the two guesses
                    if attempts == 0:
                        print("\n You run out of guesses!")
                        # addresing the user to the fail function
                        input(prompt = "<Press enter to see your result>\n")
                        fail()
                        break
                
                    # hints to the user
                    elif exact == 1:
                        print("Nope. Try again. Hint: it is the number that usually a goalkeeper has.")
                    
                    elif exact == 2:
                        print("Nope. Try again. Hint: number of people required for using a tandem.")
                        
                    elif exact == 3:
                        print("Nope. Try again. Hint: it was the perfect number according to Dante Alighieri.")
                        
                    elif exact == 4:
                        print("Nope. Try again. Hint: it is the number of members of the Beatles.")
                        
                    elif exact == 5:
                        print("Nope. Try again. Hint: it was the shirt number of Kevin Garnett.")
                        
                    elif exact == 6:
                        print("Nope. Try again. Hint: it is the number of player in a volleyball team.")
                        
                    elif exact == 7:
                        print("Nope. Try again. Hint: it is the number of Harry Potter books.")
                        
                    elif exact == 8:
                        print("Nope. Try again. Hint: if you reverse this number, it represents the infinite.")
                        
                    elif exact == 9:
                        print("Nope. Try again. Hint: it is the number of player in a baseball team.")
                        
                    elif exact == 10:
                        print("Nope. Try again. Hint: it was the shirt number of Diego Armando Maradona.")
                        
                    else:
                        # checking for bugs
                        print("Something went wrong. Try again.")
                        input(prompt = "<Press enter to restart>\n")
                        final()
                        break
                
                    # end of nested conditional
                
                else:
                    # checking for bugs
                    print("Something went wrong. Try again.")
                    input(prompt = "<Press enter if you got it.>\n")
                    final()
                    break
        
    else:
        # checking for bugs
        print("""
        Something went wrong. You need to restart the level.
        """)
        # addressing the user to the fail function
        input(prompt = "<Press enter if you got it.>\n")
        final()
    
    
# creating the winning function      
def win():
    
    # message for the winners
    print("""
Great job! Earth wins and also YOU WIN!\n
I don't know if you noticed but the correct answer was always the 'C', this is because
when I don't know the correct option I always put the 'C'. 100% working method.
    """)
    
    # giving the option to replay to the user
    replay = input(prompt = 'Would you like to play again? (Y/N)\n')
    
    # conditional statement
    if replay.lower() == 'y' or replay.lower() == 'yes':
        # the game restarts
        introduction()
        
    elif replay.lower() == 'n' or replay.lower() == 'no':
        # the game ends
        print("See you and good luck!")
        input(prompt = "<Press enter to exit>\n")
        
    else:
        # checking for bugs
        print("Something went wrong. Try again.")
        win()
        
# creating the fail function
def fail():
    
    # message for the losers
    print("""
Dude you fail this game. Your parents are already disappointed because they want you to be a DJ, not
a business analyst and now this. This is not good.\n
    """)
    
    # giving the option to replay to the user
    replay = input(prompt = 'Would you like to play again? (Y/N)\n')
    
    # conditional statement
    if replay.lower() == 'y' or replay.lower() == 'yes':
        # the game restarts
        introduction()
        
    elif replay.lower() == 'n' or replay.lower() == 'no':
        # the game ends
        print("See you and good luck!")
        input(prompt = "<Press enter to exit>\n")
        
    else:
        # checking for bugs
        print("Something went wrong. Try again.")
        fail()
        
# starting the code
introduction()
```
